```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.read_csv(school_data_to_load)
student_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data = pd.merge(student_df, school_df, on="school_name")
school_data.columns
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data.dtypes
```

```python
total_schools = len(school_data["school_name"].unique())
total_students = len(school_data["Student ID"].unique())
total_budget = sum(school_data["budget"].unique())
avg_math = round(school_data["math_score"].mean(),2)
avg_read = round(school_data["reading_score"].mean(),2)
overall_passing = (avg_math + avg_read) / 2
passing_math_count = sum(school_data["math_score"] >= 70)
passing_math_perc = round((passing_math_count / total_students)*100,2)
passing_read_count = sum(school_data["reading_score"] >= 70)
passing_read_perc = round((passing_read_count / total_students)*100,2)
district_sum_table = pd.DataFrame({"Total Schools": [total_schools], 
                                      "Total Students": [total_students],
                                      "Total Budget": [total_budget],
                                      "Average Math Score": [avg_math],
                                      "Average Reading Score": [avg_read],
                                      "Passing Math %": [passing_math_perc],
                                      "Passing Reading %": [passing_read_perc],
                                      "Overall Passing Rate %": [overall_passing]})
district_sum_table.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_summary = school_data.groupby(['school_name'])
school_name = school_summary["school_name"].unique()
school_type = school_summary["type"].unique()
school_pop = school_summary["Student ID"].count()
school_budget = school_summary["budget"].unique()
per_student_budget = school_budget / school_pop
school_avg_math = round(school_summary["math_score"].mean(),2)
school_avg_read = round(school_summary["reading_score"].mean(),2)
pass_math_score = school_data[(school_data["math_score"]>70)]
pass_math_school = pass_math_score.groupby(["school_name"]).count()["student_name"]
pass_math_perc = round((pass_math_school/school_pop)*100,2)
pass_read_score = school_data[(school_data["reading_score"]>70)]
pass_read_school = pass_read_score.groupby(["school_name"]).count()["student_name"]
pass_read_perc = round((pass_read_school/school_pop)*100,2)
sch_overall_passing = (pass_read_perc + pass_math_perc) / 2 
school_summary_table = pd.DataFrame({"School Name": school_name,
                                     "School Type": school_type,
                                     "Total Students": school_pop,
                                     "Total School Budget": school_budget,
                                     "Per Student Budget": per_student_budget,
                                     "Average Math Score": school_avg_math,
                                     "Average Reading Score": school_avg_read,
                                     "Passing Math %": pass_math_perc,
                                     "Passing Reading %": pass_read_perc,
                                     "Overall Passing Rate": sch_overall_passing
                                    })
school_summary_table.head(15)
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
top_performers = school_summary_table.sort_values("Overall Passing Rate", ascending=False)
top_performers.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
bottom_performers = school_summary_table.sort_values("Overall Passing Rate", ascending=True)
bottom_performers.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe

```python
grade_nine = student_df.loc[student_df['grade'] == '9th'].groupby('school_name')["math_score"].mean()
grade_ten = student_df.loc[student_df['grade'] == '10th'].groupby('school_name')["math_score"].mean()
grade_eleven = student_df.loc[student_df['grade'] == '11th'].groupby('school_name')["math_score"].mean()
grade_twelve = student_df.loc[student_df['grade'] == '12th'].groupby('school_name')["math_score"].mean()
math_by_grade = pd.DataFrame({"9th Grade": grade_nine,
                              "10th Grade": grade_ten,
                              "11th Grade": grade_eleven,
                              "12th Grade": grade_twelve})

math_by_grade.head(10)
```

* Perform the same operations as above for reading scores

```python
grade_nine_read = student_df.loc[student_df['grade'] == '9th'].groupby('school_name')["reading_score"].mean()
grade_ten_read = student_df.loc[student_df['grade'] == '10th'].groupby('school_name')["reading_score"].mean()
grade_eleven_read = student_df.loc[student_df['grade'] == '11th'].groupby('school_name')["reading_score"].mean()
grade_twelve_read = student_df.loc[student_df['grade'] == '12th'].groupby('school_name')["reading_score"].mean()
read_by_grade = pd.DataFrame({"9th Grade": grade_nine_read,
                              "10th Grade": grade_ten_read,
                              "11th Grade": grade_eleven_read,
                              "12th Grade": grade_twelve_read})

read_by_grade.head(10)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create Bins
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

school_summary_table["Scores by Spend"] = pd.cut(per_student_budget,bins = spending_bins,labels = group_names)
```

```python
school_math_by_spend = round(school_summary_table.groupby(["Scores by Spend"]).mean()["Average Math Score"],2)
school_reading_by_spend = round(school_summary_table.groupby(["Scores by Spend"]).mean()["Average Reading Score"],2)
school_pass_math_spend = round(school_summary_table.groupby(["Scores by Spend"]).mean()["Passing Math %"],2)
school_pass_reading_spend = round(school_summary_table.groupby(["Scores by Spend"]).mean()["Passing Reading %"],2)
overall_pass_rate_spend = (school_pass_math_spend + school_pass_reading_spend)/2
scores_by_spend = pd.DataFrame({"Average Math Score by Spending":school_math_by_spend,
                                     "Average Reading Score by Spend)":school_reading_by_spend,
                                     "% Passing Math by Spend":school_pass_math_spend,
                                     "% Passing Reading by Spend":school_pass_reading_spend,
                                     "Overall Passing rate by Spend":overall_pass_rate_spend})
scores_by_spend.head()
```
